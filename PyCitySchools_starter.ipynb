```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# school_data.count() To verify we don't have missing school data and view column names
# School ID      15
# school_name    15
# type           15
# size           15
# budget         15

# student_data.count() To verify we don't have missing student data and view column names
# Student ID       39170
# student_name     39170
# gender           39170
# grade            39170
# school_name      39170
# reading_score    39170
# math_score       39170
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools
TSch = int(float(school_data["School ID"].count()))

#Calculate the total number of students
TStu = int(float(student_data["Student ID"].count()))

#Calculate the total budget
TB = int(float(school_data["budget"].sum()))

#Calculate the average math score
AMS = round(float(student_data["math_score"].mean()),2)

#Calculate the average reading score
ARS = round(float(student_data["reading_score"].mean()),2)

#Calculate the percentage of students with a passing math score (70 or greater)
MathFilter = student_data.loc[student_data["math_score"]>=70, :]
PM = round(float(MathFilter["Student ID"].count()/TStu*100),2)

#Calculate the percentage of students with a passing reading score (70 or greater)
ReadFilter = student_data.loc[student_data["reading_score"]>=70, :]
PR = round(float(ReadFilter["Student ID"].count()/TStu*100),2)

#Calculate the percentage of students who passed math and reading (% Overall Passing)
OverallFilter = student_data.loc[(student_data["math_score"]>=70) & (student_data["reading_score"]>=70) , :]
OP = round(float(OverallFilter["Student ID"].count()/TStu*100),2)

#Create a dataframe to hold the above results
district_summary = pd.DataFrame([[TSch, TStu, TB, AMS, ARS, PM, PR, OP]], \
                                columns=["No. Schools", "No. Students", "Total Budget", "AVG Math Score", \
                                         "AVG Reading Score", "%Passing Math", "%Passing Reading", "%Overall Passing"]) 
#Display District Summary dataframe
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#List of schools with Name and Type
school_list = school_data.set_index(["school_name"])["type"]

#Calculate Total Students per school
StuPerSchool = school_data_complete["school_name"].value_counts()

#Calculate Total School Budget
BudPerSchool = school_data.set_index(["school_name"])["budget"]

#Calculate Per Student Budget
BudPerStu = BudPerSchool / StuPerSchool

#Calculate Average Math Score
AvgMathSchool = round(school_data_complete.groupby(["school_name"]).mean()["math_score"],2)

#Calculate Average Reading Score
AvgReadSchool = round(school_data_complete.groupby(["school_name"]).mean()["reading_score"],2)

#Calculate % Students Passing Math per school (70 or greater)
PassMath = school_data_complete[(school_data_complete["math_score"]>=70)]
PassMathPerS = PassMath.groupby(["school_name"]).count()["Student ID"]        
ValuePassMath = round((PassMathPerS/StuPerSchool*100),2)    

#Calculate % Students Passing Reading per school (70 or greater)
PassRead = school_data_complete[(school_data_complete["reading_score"]>=70)]
PassReadPerS = PassRead.groupby(["school_name"]).count()["Student ID"]        
ValuePassRead = round((PassReadPerS/StuPerSchool*100),2)    

#Calculate % Overall Passing (The percentage of students that passed math and reading)
PassMR = school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70), :]
PassMRPerS = PassMR.groupby(["school_name"]).count()["Student ID"]        
ValueOverall = round((PassMRPerS/StuPerSchool*100),2)   
```

```python
#Create a dataframe to hold the above results
LasEscuelas = pd.DataFrame({"School Type" : school_list,
                            "Total Students" : StuPerSchool,
                            "Total School Budget" : BudPerSchool,
                            "Per Student Budget" : BudPerStu,
                            "Avg. Math Score" : AvgMathSchool,
                            "Avg. Reading Score" : AvgReadSchool,
                            "% Passing Math" : ValuePassMath,
                            "% Passing Reading" : ValuePassRead,
                            "% Overall Passing" : ValueOverall})

#Display dataframe
LasEscuelas
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
# Scores by School Spending
# Create a table that breaks down school performances based on average Spending Ranges (Per Student). 
# Use 4 reasonable bins to group school spending. 
# Include in the table each of the following:
#        Average Math Score
#        Average Reading Score
#        % Passing Math
#        % Passing Reading
#        Overall Passing Rate (Average of the above two

# Save main criteria
grupo = "Spending Ranges (Per Student)"

# First I need to create the four reasonable bins to group values
Da_bins = [0, 585, 630, 645, 680]

# Creating names for the bins
Da_names = ["<$585", "$585-630", "$631-645", "$646-680"]

# Assigning bin name based on "Per Student Budget" values
LasEscuelas[grupo] = pd.cut(BudPerStu, bins = Da_bins, labels = Da_names)

# Calculating Average Math Score
AVG_math = round(LasEscuelas.groupby([grupo]).mean()["Avg. Math Score"],2)
AVG_read = round(LasEscuelas.groupby([grupo]).mean()["Avg. Reading Score"],2)
STUPassM = round(LasEscuelas.groupby([grupo]).mean()["% Passing Math"],2)
SUTPassR = round(LasEscuelas.groupby([grupo]).mean()["% Passing Reading"],2)
OvePassR = round(((STUPassM + SUTPassR)/2),2)

# Create dataframe
ScoreSchoolSize = pd.DataFrame({"Average Math Score"     :AVG_math,
                                     "Average Reading Score" :AVG_read,
                                     "% Passing Math"        :STUPassM,
                                     "% Passing Reading"     :SUTPassR,
                                     "% Overall Passing"     :OvePassR})

# Display DataFrame
ScoreSchoolSize
```
